```python
import pickle
import joblib
from sklearn.linear_model import  LinearRegression
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\DS_Material\FastAPI_Material\venv\Model_serialization\housing.csv')
df.head()
```

```python
df = df.iloc[:,:-1].dropna()
df.head()
```

```python
X = df.drop(columns='median_house_value')
y = df.median_house_value
```

```python
print(X.shape,y.shape)
```

## Train the model

```python
model = LinearRegression()
model.fit(X,y)
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,y_pred)
```

## Pickle File

```python
with open('model.pkl','wb') as f:
    pickle.dump(model,f)
```

```python
with open('model.pkl','rb') as f:
    picket_model = pickle.load(f)
```

```python
picket_model
```

```python
y_pred_pickle = picket_model.predict(X)
y_pred_pickle
```

```python
r2_score(y,y_pred_pickle)
```
